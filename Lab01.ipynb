```python
def artista_con_mayor_popularidad():
    archivo = open("canciones.txt", "r", encoding="utf-8")
    linea = archivo.readline()
    artista_max = ""
    popularidad_max = -1
    archivo.close()

    seguir = 1  # Control manual del ciclo

    while seguir == 1:
        archivo = open("canciones.txt", "r", encoding="utf-8")
        linea = archivo.readline()
        artista_actual = ""
        artista_encontrado = 0  # Indicador si ya encontramos un artista sin procesar

        # Buscar artista no procesado
        while linea != "" and artista_encontrado == 0:
            campo = ""
            i = 0
            barras = 0
            while i < len(linea):
                if linea[i] == '|':
                    barras += 1
                elif barras == 1:
                    campo += linea[i]
                elif barras > 1:
                    break
                i += 1

            j = 0
            nombre = ""
            while j < len(campo):
                if campo[j] == ',':
                    break
                nombre += campo[j]
                j += 1

            if len(nombre) > 0 and nombre[0] != '@':
                artista_actual = nombre.strip()
                artista_encontrado = 1
            linea = archivo.readline()
        archivo.close()

        if artista_encontrado == 0:
            seguir = 0  # Ya no hay artistas por procesar
        else:
            # Calcular promedio de popularidad
            archivo = open("canciones.txt", "r", encoding="utf-8")
            linea = archivo.readline()
            suma = 0
            cantidad = 0

            while linea != "":
                esta = 0
                k = 0
                while k + len(artista_actual) <= len(linea):
                    sub = linea[k:k+len(artista_actual)]
                    if sub == artista_actual:
                        esta = 1
                    k += 1

                if esta == 1:
                    cont_barra = 0
                    pop = ""
                    for c in linea:
                        if c == '|':
                            cont_barra += 1
                        elif cont_barra == 2:
                            pop += c
                        elif cont_barra > 2:
                            break
                    try:
                        suma += int(pop.strip())
                        cantidad += 1
                    except:
                        pass

                linea = archivo.readline()
            archivo.close()

            if cantidad > 0:
                promedio = suma / cantidad
                if promedio > popularidad_max:
                    popularidad_max = promedio
                    artista_max = artista_actual

            # Marcar artista como procesado
            archivo = open("canciones.txt", "r", encoding="utf-8")
            nuevo = open("temp.txt", "w", encoding="utf-8")
            linea = archivo.readline()

            while linea != "":
                nueva_linea = ""
                if artista_actual in linea:
                    i = 0
                    barras = 0
                    ya_marco = 0
                    while i < len(linea):
                        if barras == 1 and ya_marco == 0 and linea[i] != ',' and linea[i] != '|':
                            nueva_linea += "@"
                            ya_marco = 1
                        nueva_linea += linea[i]
                        if linea[i] == '|':
                            barras += 1
                        i += 1
                    nuevo.write(nueva_linea)
                else:
                    nuevo.write(linea)
                linea = archivo.readline()
            archivo.close()
            nuevo.close()

            archivo = open("temp.txt", "r", encoding="utf-8")
            original = open("canciones.txt", "w", encoding="utf-8")
            linea = archivo.readline()
            while linea != "":
                original.write(linea)
                linea = archivo.readline()
            archivo.close()
            original.close()

    print("🎧 El artista con mayor popularidad promedio es:", artista_max)
    print("⭐ Popularidad promedio:", round(popularidad_max, 2))
```

```python
import requests

# Reemplaza con tus credenciales de Spotify
client_id = "93177efd1fc949d7bec0bed3862aaeca"
client_secret = "6cd9ebbacefc437a9334f1d3d0504cae"
playlist_id = "5ZB7gq1VU2BwYMuQrgppg9"

# Obtener token de acceso
def get_access_token(client_id, client_secret):
    url = "https://accounts.spotify.com/api/token"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {
        "grant_type": "client_credentials",
        "client_id": client_id,
        "client_secret": client_secret
    }
    response = requests.post(url, headers=headers, data=data)
    return response.json().get("access_token")

# Guardar canciones en archivo secuencial
def guardar_canciones_archivo(playlist_id, token):
    url = f"https://api.spotify.com/v1/playlists/{playlist_id}/tracks"
    headers = {"Authorization": f"Bearer {token}"}
    with open("canciones.txt", "w", encoding="utf-8", buffering=8192) as file:
        while url:
            res = requests.get(url, headers=headers)
            data = res.json()
            for item in data["items"]:
                track = item.get("track")
                if track:
                    name = track.get("name", "N/A").replace("|", " ")
                    popularity = str(track.get("popularity", 0))
                    duration = str(track.get("duration_ms", 0))
                    artists = [artist["name"].replace("|", " ") for artist in track.get("artists", [])]
                    for artist in artists:
                        file.write(f"{name}|{artist}|{popularity}|{duration}\n")
            url = data.get("next")

# 1. Artista con más canciones
def artista_mas_canciones():
    conteo = {}
    with open("canciones.txt", "r", encoding="utf-8", buffering=8192) as file:
        for linea in file:
            _, artista, _, _ = linea.strip().split("|")
            conteo[artista] = conteo.get(artista, 0) + 1
    artista_max = max(conteo, key=conteo.get)
    print(f"🎤 Artista con más canciones: {artista_max} ({conteo[artista_max]} canciones)")

# 2. Artista con mayor índice de popularidad
def artista_mas_popular():
    popularidad = {}
    conteo = {}
    with open("canciones.txt", "r", encoding="utf-8", buffering=8192) as file:
        for linea in file:
            _, artista, pop, _ = linea.strip().split("|")
            pop = int(pop)
            popularidad[artista] = popularidad.get(artista, 0) + pop
            conteo[artista] = conteo.get(artista, 0) + 1
    promedio = {art: popularidad[art] // conteo[art] for art in popularidad}
    artista_max = max(promedio, key=promedio.get)
    print(f"🌟 Artista más popular (promedio): {artista_max} ({promedio[artista_max]})")

# 5. Canciones con duración mayor al promedio
def canciones_mayor_duracion_promedio():
    suma_dur = 0
    total = 0
    with open("canciones.txt", "r", encoding="utf-8") as file:
        for linea in file:
            _, _, _, dur = linea.strip().split("|")
            suma_dur += int(dur)
            total += 1
    promedio = suma_dur // total
    print(f"⏱ Duración promedio: {promedio} ms. Canciones que lo superan:")
    with open("canciones.txt", "r", encoding="utf-8") as file:
        for linea in file:
            nombre, artista, _, dur = linea.strip().split("|")
            if int(dur) > promedio:
                print(f"{nombre} - {artista} ({dur} ms)")

# Ejecución principal
token = get_access_token(client_id, client_secret)
guardar_canciones_archivo(playlist_id, token)
artista_mas_canciones()
artista_mas_popular()
canciones_mayor_duracion_promedio()
```
